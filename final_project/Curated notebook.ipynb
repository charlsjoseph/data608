```python
import pandas as pd
import numpy as np 
from datetime import datetime

import glob, os    

```

```python

path = r'C:/Users/Charls/Documents/CunyMSDS/Data608/AQI-DATASET/AQI-DATASET'                     # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))     # advisable to use os.path.join as this makes concatenation OS independent

df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
filtereddf1 = concatenated_df[['State Name' ,'County Name' ,'City Name' , 'Date Local' , 'AQI' , 'Parameter Name' ]]
```

```python
cleaneddf2 = filtereddf1.dropna(axis='rows') 
```

```python
cleaneddf2['Parameter Name'].unique()
```

```python

cleaneddf2.loc[cleaneddf2['Parameter Name'] == 'Carbon monoxide' , 'Parameter Name'] = 'CO'
cleaneddf2.loc[cleaneddf2['Parameter Name'] == 'Sulfur dioxide' , 'Parameter Name'] = 'SO2'
cleaneddf2.loc[cleaneddf2['Parameter Name'] == 'Nitrogen dioxide (NO2)' , 'Parameter Name'] = 'NO2'
cleaneddf2.loc[cleaneddf2['Parameter Name'] == 'Ozone' , 'Parameter Name'] = 'O3'
```

```python
cleaneddf2['Parameter Name'].unique()
```

```python
cleaneddf2.head(10)
```

```python
spreadDf3=pd.pivot_table(cleaneddf2,index=['State Name','County Name' ,'City Name' , 'Date Local' ],columns='Parameter Name',values='AQI')
```

```python
spreadDf3 = spreadDf3.fillna(0)
```

```python
spreadDf3.head()
spreadDf3 = spreadDf3.reset_index()
```

```python
spreadDf3 = spreadDf3.rename(columns = {'State Name' : 'State','County Name' : 'County', 'City Name' : 'City' , 'CO' : 'CO AQI' , 'NO2' : 'NO2 AQI' , 'O3' : 'O3 AQI' , 'SO2' : 'SO2 AQI' })
```

```python
spreadDf3
```

```python
spreadDf3['AQI'] = spreadDf3[['NO2 AQI' , 'O3 AQI', 'SO2 AQI' ,'CO AQI']].values.max(1)
spreadDf3['year'] = spreadDf3['Date Local'].apply(lambda x : datetime.strptime(x, '%Y-%m-%d').year)
spreadDf3['month'] = spreadDf3['Date Local'].apply(lambda x : datetime.strptime(x, '%Y-%m-%d').month)
```

```python
spreadDf3.head(16)
```

```python


filteredDFYearly = spreadDf3.groupby(['State', 'year']) \
                        .max().reset_index() \
                        .drop(['month'], axis =1)
                    

filteredDFMonthly = spreadDf3.groupby(['State', 'year', 'month']).max().reset_index()

```

```python
from functools import reduce
def countagg(series):
    map(lambda x : 1 if x> 200 else 0, series)
    return reduce(lambda x , y : x + y, series)

```

```python
df2_montly =spreadDf3.groupby(['State', 'County' , 'year', 'month']).agg({'O3 AQI': countagg,\
                                                        'CO AQI': countagg, \
                                                        'NO2 AQI': countagg, \
                                                        'SO2 AQI': countagg}).reset_index() 

df2_state =spreadDf3.groupby(['State', 'County']).agg({'O3 AQI': countagg,\
                                                        'CO AQI': countagg, \
                                                        'NO2 AQI': countagg, \
                                                        'SO2 AQI': countagg}).reset_index() 
```

```python
df2_state['total'] = df2_state['O3 AQI'] + df2_state['CO AQI'] + df2_state['NO2 AQI'] + df2_state['SO2 AQI']
```

```python
df2_state = df2_state.sort_values(by=['total'], ascending=False)
 
```

```python
df2_state = df2_state.rename(columns ={'O3 AQI':'O3-AQI','CO AQI':'CO-AQI' , 'NO2 AQI': 'NO2-AQI', 'SO2 AQI' :'SO2-AQI' })


```

```python
df2_state.to_csv("C:/Users/Charls/Documents/CunyMSDS/Data608/solutions/finalproject/application/data/topcounties.csv", index=False)
```

```python
df2_montly = df2_montly.rename(columns ={'O3 AQI':'O3-AQI','CO AQI':'CO-AQI' , 'NO2 AQI': 'NO2-AQI', 'SO2 AQI' :'SO2-AQI' })
df2_montly['total'] = df2_montly['O3-AQI'] + df2_montly['CO-AQI'] + df2_montly['NO2-AQI'] + df2_montly['SO2-AQI']
df2_montly.to_csv("C:/Users/Charls/Documents/CunyMSDS/Data608/solutions/finalproject/application/data/monthly_all_states.csv", index=False)

```
