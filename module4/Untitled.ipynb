```python
import pandas as pd
import numpy as np
```

```python
soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=*' +\
        '&$where=health  is not null &$limit=80000').replace(' ', '%20')
treesall = pd.read_json(soql_url)
treesall.shape[0]
```

```python
trees_count = treesall.groupby(["steward", "health" ]).tree_id.agg(['count']).reset_index()
trees_count['prop'] = trees_count['count'].transform(lambda x : x/x.sum()*100)
```

```python
trees_count.to_csv("contigency_data.csv")
```
